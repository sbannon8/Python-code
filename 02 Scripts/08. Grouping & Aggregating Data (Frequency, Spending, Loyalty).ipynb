**------  LOYALTY FLAG  ---------------------**

Loyalty categories are as follows:
- If max orders made by user > 40 : "Loyal customer"
- If max orders > 10 and <= 40: "Regular customer"
- If max orders <= 10: "New customer"

```python
# creates a new col that shows a user id's total num of orders
ords_prods['max_order'] = ords_prods.groupby(['user_id'])['order_number'].transform(np.max)
ords_prods.head(5)
```

```python
# assign loyalty flags in a new col caled loyalty-flag
# Loyal
ords_prods.loc[ords_prods['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Regular
ords_prods.loc[(ords_prods['max_order'] > 10 ) & (ords_prods['max_order'] <= 40), 'loyalty_flag'] = 'Regular customer'
```

```python
# New
ords_prods.loc[ords_prods['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Number breakdown of each category
ords_prods['loyalty_flag'].value_counts(dropna = False)
```

```python
# (QUESTION 5)
# show average price of products purchased by loyalty flag
ords_prods.groupby('loyalty_flag').agg({'prices':['mean']})

# spend about the same per product
```

**------   SPENDING FLAG   ---------------------**

Spending categories are as follows:
- If mean price of products purchased by user >= 10 : 'High spender'
- If mean price < 10 : 'Low spender'

```python
# (QUESTION 6)
# create new col that shows a user's avg per product
ords_prods['avg_spent'] = ords_prods.groupby(['user_id'])['prices'].transform(np.mean)
ords_prods.head(5)
```

```python
# assign labels for newly created avg per product
# High spender
ords_prods.loc[ords_prods['avg_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Low spender
ords_prods.loc[ords_prods['avg_spent'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods['spending_flag'].value_counts(dropna = False)
```

**------   FREQUENCY FLAG   ---------------------**

Frequency categories are as follows:
- If median days_since_prior_order > 20 : 'Non-frequent customer'
- If median days > 10 & <= 20 : 'Regular customer'
- If median days <= 10 : 'Frequent customer'

```python
# (QUESTION 7)
# create new col that shows a user's median days_since_prior_order
ords_prods['med_since_prior'] = ords_prods.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
ords_prods.head(5)
```

```python
# assign labels for newly created med days since
# Non-frequent
ords_prods.loc[ords_prods['med_since_prior'] > 20, 'frequency_label'] = 'Non-frequent customer'
```

```python
# Regular customer
ords_prods.loc[(ords_prods['med_since_prior'] > 10) & (ords_prods['med_sinceprior'] <= 20), 'frequency_label'] = 'Regular customer'
```

```python
# Frequnt customer
ords_prods.loc[ords_prods['med_since_prior'] <= 10, 'frequency_label'] = 'Frequent customer'
```

```python
ords_prods['frequency_label'].value_counts(dropna = False)
```
