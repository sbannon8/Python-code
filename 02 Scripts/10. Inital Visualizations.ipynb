## Importing Libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

## Importing Data

```python
# main folder
path = r'C:\Users\steve\Documents\11.24 Instacart Basket Analysis'

# ords_prods_cust
ords_prods_cust = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_cust.pkl'))
```

## Visualizations

**Histogram of Order Hour of Day**

```python
hour_hist = ords_prods_cust['order_hour_of_day'].plot.hist(bins=24)
plt.xlabel('Hour of Day')
plt.ylabel('Order Frequency')
plt.title('Order Hour of Day Histogram')
```

The histogram shows that majority of orders happen between hours 9 - 16, with a build up before and steady descrease after. Very few orders happen during hours 24 and 0 - 5. The max orders occurs at hour 9 and the minimum at hour 3.

**Bar Chart of Loyalty Flag**

```python
loyalty_bar = ords_prods_cust['loyalty_flag'].value_counts().sort_index().plot.bar()
plt.xlabel('Loyalty Status')
plt.ylabel('Number of Customers')
plt.title('Customer Loyalty')
```

**Difference in Expenditure Depending on Hour of Day**

01. Sample Data

```python
# assign a random number to every data entry
np.random.seed(4)
# create a subset that includes any entry that was randomly assigned a 0.7 or lower
dev = np.random.rand(len(ords_prods_cust)) <= 0.7
# store 70% in big_sample
big_sample = ords_prods_cust[dev]
# store 30% in samll_sample
small_sample = ords_prods_cust[~dev]
```

02. Create DF

```python
df = small_sample[['order_hour_of_day', 'prices']]
```

03. Line Chart

```python
spentHourly_line = sns.lineplot(data = df, x = 'order_hour_of_day', y = 'prices').set(title = 'Expenditure by Hour of Day')
```

**Customer Age vs Number of Dependents**

```python
# create df with a sample of the 2 needed col
df_2 = small_sample[['age', 'dependents']]
```

```python
# create line
ageNumDependents_line = sns.lineplot(data = df_2, x = 'age', y = 'dependents').set(title = 'Customer Age vs. Number of Dependents')
```

The line chart shows that there is no connection between customer age and number of dependents.  All ages have an average of 1 - 2 dependents.

**Age vs Spending Power(Income)**

```python
# creae df with sample of the 2 needed col
df_3 = small_sample[['age', 'income']]
```

```python
# create scatterplot
ageIncome_scatter = sns.scatterplot(x = 'age', y = 'income', data = df_3).set(title = 'Age vs. Spending Power(Income)')
```

The scatterplot shows that 18 - 40 year olds have an average of about 100,000. After 40, there is an evident increase in spending power with the average raising to about 200,000.
